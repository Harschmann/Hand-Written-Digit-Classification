```python
#Import the necessary libraries
import tensorflow as tf
from tensorflow import keras
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Load training and testing MNIST dataset from keras
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#Display training set shape 
#60,000 images, each being 28 x 28 pixels
X_train.shape       
```

```python
#Scale the pixels between 0 and 1
X_train = X_train / 255
X_test = X_test / 255
```

```python
#Flatten 2D array into 1D array
#Convert each image from 28 x 28 pixels to 784 pixels
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)

#X_train_flattened.shape       #Use this to see the flattened image
```

```python
X_train_flattened[0]        #Show all the 28 x 28 = 784 pixels as 1D array
```

```python
#Create One layer NN

'''
Keras.Sequential allows creating stack of layers
keras.layers.dense allows to connect each in one layer to connect with each neuron in second layer
'''
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

'''
Use 'adam' optimizer for optimizing the model
Use 'spare_categorical_crossentropy'
Use 'accuracy' as metric
'''
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

#Train the model
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
#Evaluate model accuracy on testing dataset
model.evaluate(X_test_flattened, y_test)
```

```python
#Show first testing image
plt.matshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
#Predict the image
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
#Print confusion matrix
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
#Visualize confusion matrix
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#Create two layer NN i.e. add one hidden layer

'''
Keras.Sequential allows creating stack of layers
keras.layers.dense allows to connect each in one layer to connect with each neuron in second layer
Flatten with the help of keras
'''
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(77, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

'''
Use 'adam' optimizer for optimizing the model
Use 'spare_categorical_crossentropy'
Use 'accuracy' as metric
'''
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

#Train the model
model.fit(X_train, y_train, epochs=10)
```

```python
#Evaluate model accuracy on testing dataset
model.evaluate(X_test, y_test)
```
